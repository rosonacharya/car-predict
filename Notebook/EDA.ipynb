```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the data
data = pd.read_csv('../data/datacar.csv')
print(data.head())
```

```python
#seprataing catorgical and numerical data
cat_data = data.select_dtypes(include=['object']).columns
num_data = data.select_dtypes(exclude=['object']).columns
```

# Univariate Analysis

```python
#univarient analysis
for col in num_data:
    plt.figure(figsize=(3,3))
    sns.histplot(data[col], kde=True, bins=20)
    plt.title(col)
```

```python
#subplots
fig, axes = plt.subplots(2,3,  figsize=(10, 10))
axes = axes.flatten()
for i, col in enumerate (num_data):
    sns.histplot(data[col], kde=True, bins=20, ax=axes[i])
    axes[i].set_title(col)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#count plot for categorical data
fig, axes = plt.subplots(2, 3, figsize=(10,10))
axes = axes.flatten()
for i, cat in enumerate(cat_data):
    sns.countplot(x=data[cat], ax=axes[i])
    axes[i].set_title(cat)
    axes[i].set_xlabel(cat)
    axes[i].set_ylabel('Frequency')
fig.delaxes(axes[-1])
plt.tight_layout()
plt.show()
```

## Bivarent analysis

```python
## Bivarent analysis
sns.histplot(data=data, x='year' ,hue='owner', multiple="dodge", bins=20)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=data, x='year', hue='fuel', multiple="dodge", bins=20)
plt.show()
```
